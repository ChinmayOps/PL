```python
temp_df = df_heart[['age','cp', 'trtbps','chol','fbs','oldpeak','output']]
plt.figure(figsize=(15,10))
sns.pairplot(temp_df,hue="output")
plt.title("Looking for Insites in Data")
plt.legend("HeartDisease")
plt.tight_layout()
plt.plot()
```

```python
plt.figure(figsize=(15,10))
for i,col in enumerate(temp_df.columns,1):
 plt.subplot(4,3,i)
 plt.title(f"Distribution of {col} Data")
 sns.histplot(df_heart[col],kde=True)
 plt.tight_layout()
 plt.plot()
```

```python
sns.lineplot(df_heart, x='age', y='thalachh')
plt.show()
```

```python
sns.jointplot(data=df_heart, x='age', y='chol', kind='scatter')
plt.suptitle("Jointplot of Age vs Cholesterol", y=1.02)
plt.show()
```
